**Seasonal Ratings**

```python
def cour_seasons(date):
    if date.month in [1,2,3]:
        return "Winter Cour"
    elif date.month in [4,5,6]:
        return "Spring Cour"
    elif date.month in [7,8,9]:
        return "Summer Cour"
    else:
        return "Fall Cour"

anime_data['Season Cour'] = anime_data['Start Date'].apply(cour_seasons)
```

```python
seasonal_show_ranking = anime_data[['Name','Score', 'Season Cour']].copy()
seasonal_show_ranking
seasonal_show_ranking.boxplot(column='Score', by='Season Cour')
plt.xlabel('Seasonal Cour')
plt.ylabel('Score')
plt.title('Score Distribution by Seasonal Cour')
plt.xticks(rotation=45)
plt.show()
```

**Box and Whisker plot for seasonal ranking data** so after some research this seemed the best bet for this data. the position and size of the box on the Y Axis is the overall range of the scores for each season. each box length is based on the overall spread of the scores in each season. so we can see that fall and winter are longer boxes more shows and score variance. the green line is the median score for that season of all shows released in that 3 month bracket. the black horizonatal line denotes where the highest score ends up for that season and beyond it; statiscal outliers, which are represented by dots on this visual.

**Findings**

because this data set is made up of the top 2000 ranked anime on MAL there is a narrower range for score variance. to begin with our lowest ranked score to crack the top 2000 is 7.50, which is slightly lower than id think would be represented here. 2.There is some variation among the median scores. so there is some varation albeit not the most significant between quality in seasons.
